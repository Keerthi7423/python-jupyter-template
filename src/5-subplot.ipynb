### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.arange(1, 6)  # x values: [1, 2, 3, 4, 5]
y = x * 2  # y values: [2, 4, 6, 8, 10]
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line chart in the first subplot
axes[0, 0].plot(x, y, marker='o', linestyle='-', color='blue', label='y = 2x')
axes[0, 0].set_title("Line Chart")
axes[0, 0].set_xlabel("x values")
axes[0, 0].set_ylabel("y values")
axes[0, 0].legend()
axes[0, 0].grid(True, linestyle='--', alpha=0.6)

# Bar chart in the second subplot
axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axes[0, 1].set_title("Bar Chart")
axes[0, 1].set_xlabel("Categories")
axes[0, 1].set_ylabel("Values")

# Scatter plot in the third subplot
axes[1, 0].scatter(x, y, color='orange')
for i, j in zip(x, y):
    axes[1, 0].text(i, j, f'({i}, {j})', fontsize=8, ha='right', va='bottom')
axes[1, 0].set_title("Scatter Plot")
axes[1, 0].set_xlabel("x values")
axes[1, 0].set_ylabel("y values")

# Pie chart in the fourth subplot
axes[1, 1].pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    colors=['gold', 'lightblue', 'lightgreen', 'coral'],
    explode=(0.1, 0, 0, 0)
)
axes[1, 1].set_title("Pie Chart")

# Adjust layout for better spacing
plt.tight_layout()

# Display the plots
plt.show()
```
